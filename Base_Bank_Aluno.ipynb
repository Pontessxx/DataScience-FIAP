## imports

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics


```

```python
base = pd.read_csv('./csv/bank-full.csv', delimiter=';')
```

## Base (DataFrame principal)

```python
base
```


### Retirando variaveis categoricas nao dummificadas

```python
# Drop colunas categóricas
base_drop = base.drop(columns=["job","marital", "education","default", "housing", "loan", "contact", "month", "poutcome"], axis=1)
base_drop
```

### Dummy das variaveis categoricas

```python
# Obter colunas dummy apenas para as categorias relevantes
categorica = pd.get_dummies(base[["job","marital", "education","default", "housing", "loan", "contact", "month", "poutcome"]], dtype=int)
categorica
```

```python
base_nova = pd.concat([base_drop, categorica], axis=1)
base_nova
```

## Separando em treino teste

```python
X = base_nova.drop(['y'],axis=1)
y = base_nova['y']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.35, random_state=42)
```

```python
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
```

```python
print(f'{round(model.score(X_test, y_test) * 100,2 )}%')
```
