```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
base = pd.read_csv('./csv/mobile.csv')
```

```python
base
```

```python
base.columns
```

```python
base.iloc[:, 0:20]
```

```python
categorias = base[["blue", "dual_sim", "four_g", "three_g", "touch_screen", "wifi"]]
categorias = pd.get_dummies(categorias, columns=["blue", "dual_sim", "four_g", "three_g", "touch_screen", "wifi"],dtype=int)
categorias
```

```python
base2 = base.drop(columns=["blue", "dual_sim", "four_g", "three_g", "touch_screen", "wifi"])
base2
```

```python
x = base2.drop("price_range", axis=1)
y = base["price_range"]

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.3, random_state=1,stratify=y)
```

```python
model = DecisionTreeClassifier()
model.fit(x_treino, y_treino)
```

```python
model.acuracia = model.score(x_teste, y_teste)
```
